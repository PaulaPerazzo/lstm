# importing libs

```python
import yfinance
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

# Getting historical and .csv file

```python
ticker = 'AMZN'

# get historical data
amazon_data = yfinance.download(ticker, start='2000-01-01', end='2023-12-31', interval='1d')
print(amazon_data)
```

```python
# turning the data into a csv file
amazon_data.to_csv('amazon_data.csv')
```

```python
amazon_df = pd.read_csv('./data/amazon_data.csv')
print(amazon_df.head())
```

# Some info about the dataset

```python
print(amazon_df.info())
```

```python
# get the summary statistics
amazon_df.describe()
```

# Graphic visualization

```python
### boxplot
sns.boxplot(data=amazon_df[['Adj Close', 'Open', 'High', 'Low']])
```

```python
### line plot
amazon_df.set_index('Date', inplace=True)
sns.lineplot(data=amazon_df[['Adj Close', 'Open', 'High', 'Low']]).xaxis.set_major_locator(mdates.YearLocator(2))

plt.xticks(rotation=45)

plt.title('Amazon OHLVC data from 2000 to 2023')
plt.xlabel('Date')
plt.ylabel('Value in USD')

# Mostra o gráfico
plt.show()
```
